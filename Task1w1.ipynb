```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df = pd.read_csv('coursera.csv')
```

```python
print(df)
```

```python
df['course_rating'].fillna(df['course_rating'].mean(),inplace=True)
print(df)
```

```python
df.info()
```

```python
df['skill_gain'].fillna(df['skill_gain'].mode()[0],inplace=True)
print(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['course_review'])
df.info()
```

```python
df.dropna(subset=['course_type','duration'],inplace=True)
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()
df['course_type_label']=le.fit_transform(df['course_type'])
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='course_type', data=df)
plt.title('Distribution of Courses by Type')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df)
```

```python
sns.lineplot(x='course_type',y='course_rating', data=df)
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=df['course_rating'], shade=True)
plt.title('Density Plot of Course Rating')
plt.xlabel('Course rating')
plt.ylabel('Density')
plt.show()
```

```python
df.to_csv('pp_coursera.csv',index=False)
```
